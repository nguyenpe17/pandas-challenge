```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load
pymoli = "resources/04-Pandas_HW_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
pymoli_data = pd.read_csv(pymoli,encoding='utf-8')

# displays the data set
pymoli_data
```

## Player Count

* Display the total number of players


```python
# Calculating total players
total_players = len(pymoli_data["SN"].unique())

# Creating a data frame w/total_players count
player_count = pd.DataFrame({"Total players":total_players},index = [0])
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Purchasing Analysis (Total)

# Calculate the count of unique items
unique_items_count = len(pymoli_data["Item ID"].unique())

# calculate the average price of the purchase 
average_price = pymoli_data["Price"].mean()

# caluclate the total number of purchases made 
number_purchases = pymoli_data["Purchase ID"].count()

# calculate the total revenue
total_revenue = pymoli_data["Price"].sum()

#Summary data frame = unique items, average price, number of purchases, total revenue
Purchasing_Analysis = pd.DataFrame({"Number of unique items":unique_items_count,"Average Price":average_price,
                                    "Number of Purchases":number_purchases,"Total Revenue":total_revenue}, index=[0])

Purchasing_Analysis["Average Price"] = Purchasing_Analysis["Average Price"].map("${:.2f}".format)

Purchasing_Analysis["Total Revenue"] = Purchasing_Analysis["Total Revenue"].map("${:,.2f}".format)

Purchasing_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Gender Demographics

gender_demo = pymoli_data.groupby("Gender")
other_members = gender_demo["SN"].nunique()
percentage_of_players = (other_members/total_players)*100

# Summary data = Percentage and Count of Male, Female, Other/Non-Disclosed
gender_summary = pd.DataFrame({"Percentage of Players": percentage_of_players,"Total Counts": other_members})

gender_summary["Percentage of Players"] = gender_summary["Percentage of Players"].map("{:.2f}".format)

gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchasing Analysis (Gender)

# Group each category using a groupby
gender_demo = pymoli_data.groupby("Gender")
purchase_count = gender_demo["Age"].count()
avg_purchase = gender_demo["Price"].mean()
total_purchase_value = gender_demo["Price"].sum()
avg_purchase_person =(gender_demo["Price"].sum()/other_members)

# Summary data frame = Purchase Count, Avg Purchase Price, Total Purchase per person (by gender)
summary_df = pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":avg_purchase,
                           "Total Purchase value":total_purchase_value,"Average Purchase Total per person":avg_purchase_person})

summary_df["Average Purchase Price"] = summary_df["Average Purchase Price"].map("${:.2f}".format)
summary_df["Total Purchase value"]  = summary_df["Total Purchase value"].map("${:,.2f}".format)
summary_df["Average Purchase Total per person"] = summary_df["Average Purchase Total per person"].map("${:.2f}".format)

summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Age Demographics

age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice and categorize the existing players
pymoli_data["Age Group"] = pd.cut(pymoli_data["Age"],age_bins, labels=group_names)

pymoli_data

pymoli_age_group = pymoli_data.groupby("Age Group")

# Calculate total count and percenatge of players by age category
other_members = pymoli_age_group["SN"].nunique()

Percent_of_players = (other_members/total_players)*100

# Data Frame for calculations 
age_demographics = pd.DataFrame({"Percentage of players":Percent_of_players,"Total count":other_members})

# Display Age Demographics table

age_demographics.index.name = None

age_demographics["Percentage of players"] = age_demographics["Percentage of players"].map("{:.2f}".format)

age_demographics

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Purchasing Analysis (Age)

# Calculations for purchase count, avg. purchase price, total purchase value , average purchase per person
purchase_count = pymoli_age_group["Age"].count()
avg_purchase = pymoli_age_group["Price"].mean()
total_purchase_value = pymoli_age_group["Price"].sum()
avg_purchase_person =(pymoli_age_group["Price"].sum()/other_members)

# Data Frame for calculations 
Purchase_analysis_age = pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":avg_purchase,
                           "Total Purchase value":total_purchase_value,"Average Purchase Total per person":avg_purchase_person})

# Display & format calculations
Purchase_analysis_age["Average Purchase Price"] = Purchase_analysis_age["Average Purchase Price"].map("${:.2f}".format)

Purchase_analysis_age["Total Purchase value"]  = Purchase_analysis_age["Total Purchase value"].map("${:,.2f}".format)

Purchase_analysis_age["Average Purchase Total per person"] = Purchase_analysis_age["Average Purchase Total per person"].map("${:.2f}".format)

Purchase_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Top Spenders

# Set variable and groupby for screen name
pymoli_SN_group = pymoli_data.groupby("SN")

# Calculations for purchase count, avg. purchase price, total purchase value , average purchase per person
purchase_count = pymoli_SN_group["Age"].count()
avg_purchase = pymoli_SN_group["Price"].mean()
total_purchase_value = pymoli_SN_group["Price"].sum()

# Data Frame for calculations 
Top_spenders = pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":avg_purchase,
                           "Total Purchase value":total_purchase_value})

# Sort the total purchase value column in descending order 
Top_spenders = Top_spenders.sort_values("Total Purchase value", ascending = False)
 
# Display + format result    
Top_spenders["Average Purchase Price"] = Top_spenders["Average Purchase Price"].map("${:.2f}".format)

Top_spenders["Total Purchase value"]  = Top_spenders["Total Purchase value"].map("${:,.2f}".format)

Top_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Most Popular Items

# Retrieve the Item ID, Item Name, and Item Price columns (using .loc)  
popular_df = pymoli_data.loc[:,["Item ID","Item Name","Price"]]

# Group by Item ID and Item Name. 
popular_group = pymoli_data.groupby(["Item ID","Item Name"])

# Calculations for purchase count, item price, and total purchase value
purchase_count = popular_group["Age"].count()
avg_purchase = popular_group["Price"].mean()
total_purchase_value = popular_group["Price"].sum()

# Data Frame for calculations 
most_popular_items = pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":avg_purchase,
                                   "Total Purchase value":total_purchase_value})

# Sort the purchase count  column in descending order 
most_popular_items = most_popular_items.sort_values("Purchase count", ascending = False)

# Display & format results 
most_popular_items["Total Purchase value"]  = most_popular_items["Total Purchase value"].map("${:,.2f}".format)

most_popular_items["Average Purchase Price"] = most_popular_items["Average Purchase Price"].map("${:.2f}".format)
 
most_popular_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order
